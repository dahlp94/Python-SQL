```python
import pandas as pd
from pandasql import sqldf
from pandasql import load_births

births_db = load_births()
```

```python
births_db.columns
```

```python
print(sqldf("SELECT * FROM births_db"))
```

# A query can span mutiple lines

```python
q = """
    SELECT
        date(date) as DOB,
        sum(births) as "Total Births"
    FROM
        births_db
    GROUP BY
        date(date)
    LIMIT 10;
  """
print(sqldf(q))
```

```python
print(sqldf(q, globals()))
```

```python
def pysqldf(q):
    "add this to your script if you get tired of calling locals()"
    return sqldf(q, globals())
```
